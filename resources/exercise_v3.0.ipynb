3. 데이터 분석을 해 봅시다. 우리는 a.수치를 통한 데이터 분석, b.시각화를 통한 데이터 분석을 할 예정입니다.

b. 시각화를 통한 데이터 분석을 클래스로 만드세요. 클래스를 생성할 때, 괄호( ) 안에 a.의 내용을 추가하여
b에서 a클래스의 자원(리소스)를 사용할 수 있도록 합시다. [이를 클래스 상속이라고 합니다.]

4. 첫 번째 클래스(a)에서 데이터 분석을 하는 함수를 정의하고, 실행시켜 봅시다.
예시에서는 Covid 함수를 정의하고, 여기서 처음 Covid()클래스를 사용(호출)하면 데이터셋을 불러옵니다.
self.data 명령어로 데이터를 사용할 수 있습니다.

```python
class Covid:
    # Covid 클래스 처음 시작시
    # 데이터셋을 불러오는 부분.
    def __init__(self):
        self.data = pd.read_csv('country_wise_latest.csv')
    
    #아래는 Covid클래스에서 실행할 수 있는 기능입니다.
    
    #사망자가 가장 적은 10개의 국가 출력
    def small_deaths(self):
        print(self.data.nsmallest(10,'Deaths')) #Note which is equal to no deaths at all.
```


위와 같이 def small_deaths(self): 함수를 클래스 내부에 정의하고, 
안에 분석기능을 위치시켜 봅시다. Covid.small_deaths() 함수는 사망자가 가장 적은 10개의 국가를 출력하는 기능입니다.

5. 두 번째 클래스(b)에서 데이터 시각화를 하는 함수를 정의하고, 실행시켜 봅시다.

Visualise(Covid) 클래스에서 괄호 안에 (Covid) 클래스를 입력해 줌으로서,
Visualise 클래스에서는 Covid클래스에서 사용하던 데이터 및 함수를 그대로 사용할 수 있습니다.
confirmed_by_country() 함수는 코로나 감염자가 가장 많은 20개의 국가를 도표로 출력하는 기능입니다.

Visualise 함수에서 각종 다양한 도표를 그려 보세요!!

```python
class Visualise(Covid):
    def pie_diagram(self):
        plt.figure(figsize=(10,15))
        self.data['Country/Region'].head(20).value_counts().plot.pie(autopct="%1.1f%%")     #Note use of pie diagram.
        plt.title('Top 20 Country With Max Cases',fontsize=25)                         #Note Labelling of diagram
        plt.ylabel('Countries',fontsize=10)
        plt.show()

    def confirmed_by_country(self):
        plt.style.use('Solarize_Light2')
        plt.figure(figsize=(15,10))
        sns.countplot(data=data.head(20),y='Confirmed')                    #Note here we are using countplot as sns which is an in-nuilt library in python that uses mtplotlib underneath to plot graphs.
        plt.xticks(rotation='vertical')                                    #Note use of ticks for a better view of diagram.
        plt.title('Top Confirmed Cases By Country',fontsize=25)            #Note labelling the data
        plt.xlabel('Confirmed Cases')                               
        plt.show()
```

### 4. 아래와 같이 두 개의 클래스를 정의합니다.

첫 번째 클래스는 Covid로 정의하고, 여기서 코로나19 데이터셋을 불러와 각종 수치를 계산해 봅시다.

먼저 country_wise_latest.csv 파일을 불러옵니다.

첫 번째 Covid 클래스에서는 self.data를 출력하면

country_wise_latest.csv 의 내용이 파이썬으로 들어온 뒤

해당 내용을 출력합니다.

```python
class Covid:
    # Covid 클래스 처음 시작시
    # 데이터셋을 불러오는 부분.
    def __init__(self):
        self.data = pd.read_csv('country_wise_latest.csv')
    
    #아래는 Covid클래스에서 실행할 수 있는 기능입니다.
    
    # (1) 사망자가 가장 적은 10개의 국가 출력
    def small_deaths(self):
        print(self.data.nsmallest(10,'Deaths')) #Note which is equal to no deaths at all.
    # (2) 데이터셋의 앞부분을 출력해주는 기능
    def head(self):
        print(self.data.head())        
    # (3) 데이터에 결측치(missing value)가 있는지 체크
    def is_missing_dataset(self):
        print(self.data.isnull().sum())
    # (4) 가장 사망자가 많았던 국가
    def max(self):
        print(self.data.max())
    # (5) 가장 사망자가 적었던 국가
    def min(self):
        print(self.data.min())
    # (6)
    def value_count(self):
        print(self.data.head(10).value_counts())
    # (7)
    def describe(self):
        print(self.data.describe())
```

```python
# Covid 형태의 클래스를 정의합니다.
c = Covid()
# Covid 클래스 내에 처음에 불러온 country_wise_latest.csv를
# c.data 명령어로 출력해 봅시다.
c.data
```

(1) small_deaths()는 사망자가 가장 적었던 10개의 국가를 출력합니다.

```python
c.small_deaths()
```

(2) head()는 데이터셋의 가장 앞부분 5개의 레코드를 출력합니다.

```python
c.head()
```

(3) missing_dataset()은 데이터 열마다 결측치(오류 또는 값이 없는 경우)의 수를 체크하는 기능입니다.
c.is_missing_dataset()을 실행하여 결측치 수를 계산해 봅시다!!

```python
c.is_missing_dataset()
```

(4) c.max는 감염자가 가장 많았던 국가를 출력합니다.

```python
c.max()
```

(5) min()은 가장 사망자가 적었던 국가를 출력합니다.

```python
c.min()
```

small_deaths() 기능은 사망자가 가장 적은 국가를 차례대로 10개 출력하는 기능입니다. 실행해 봅시다.

```python
c.small_deaths()
```

(6) value_count() 기능은 사망자가 가장 많았던 10개의 국가의 통계를 출력합니다.

```python
c.value_count()
```

(7) describe() 기능은 데이터의 통계지표(평균, 표준편차, 중앙값, 사분위값)을 출력해주는 기능입니다.

### 5. 두 번째 클래스인 Visualise( ) 를 정의해 봅시다.

이 클래스에서는 괄호( ) 안에 Covid 클래스를 포함시켜서 처음부터 생성합니다.

Covid 클래스를 부모 객체, Explore 클래스를 자식 객체라고 하는데, 

> class Explore(Covid) 
</br>

와 같이 정의하면

Explore 내에서 Covid 클래스 내부의 리소스(데이터 및 기타 정의된 내용)을 그대로 불러와서 쓸 수 있다고 하여, 상속(inheritance)이란 용어를 사용합니다.



matplotlib 라이브러리 중,
도표의 종류가 무엇이 있는지 살펴봅시다.

```python
print(plt.style.available)
```

```python
class Visualise(Covid):
    def pie_diagram(self):
        plt.figure(figsize=(10,15))
        self.data['Country/Region'].head(20).value_counts().plot.pie(autopct="%1.1f%%")     #Note use of pie diagram.
        plt.title('Top 20 Country With Max Cases',fontsize=25)                         #Note Labelling of diagram
        plt.ylabel('Countries',fontsize=10)
        plt.show()

    def confirmed_by_country(self):
        plt.style.use('Solarize_Light2')
        plt.figure(figsize=(15,10))
        sns.countplot(data=data.head(20),y='Confirmed')                    #Note here we are using countplot as sns which is an in-nuilt library in python that uses mtplotlib underneath to plot graphs.
        plt.xticks(rotation='vertical')                                    #Note use of ticks for a better view of diagram.
        plt.title('Top Confirmed Cases By Country',fontsize=25)            #Note labelling the data
        plt.xlabel('Confirmed Cases')                               
        plt.show()

    def recovered_countries(self):
        plt.style.use('seaborn-poster')
        plt.figure(figsize=(15,5))
        plt.title("Top 10 Recovered Countries ")     #Note adding title to the data.
        plt.xlabel("Count")                          #Note labelling of x-axis.
        plt.ylabel('Countries Value')
        data["Recovered"].value_counts().head(20).plot(kind="barh",color="g")      #Note adding color to the horizontal bars.
        plt.show()
        
    def top_20_death(self):
        plt.figure(figsize=(10,15))       #Note the figure size
        data['Deaths'].head(20).value_counts().plot.pie(autopct="%1.1f%%")    #Note use of pie plot
        plt.title('Top 20 Deaths With Countries',fontsize=25)                 #Note leabelling the diagram
        plt.xlabel('Countries',fontsize=15)
        plt.show()
# 객체 v를 아래와 같이 정의합니다.
# v를 정의할 때, 위에 classViualise(Covid) 와 같이 정의하면,
# Covid 객체에 포함된 데이터도 같이 전달되어 Visualise() 객체에서 사용할 수 있어요!!
# 이와 같은 내용을 상속(inheritance)라고 합니다.
v = Visualise()
```

아래와 같이 v.pie_diagram() 명령어를 사용하여 도표를 그려 봅시다.

```python
v.pie_diagram()
```

```python
v.confirmed_by_country()
```

```python
v.recovered_countries()
```

```python
v.top_20_death()
```
