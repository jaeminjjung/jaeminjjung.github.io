```python
# ëŒ€ì²´í•œ corpus
# ìˆ˜ë™ìœ¼ë¡œ ì…ë ¥í•œ corpus ë‚´ìš©
corpus = [
'ì–´ë‘ìš´ ê³³ì—ì„œ 2ê°œì›” ì´ìƒ ë°œíš¨ì‹œí‚µë‹ˆë‹¤.',
 'ì‹±ì–´ì†¡ë¼ì´í„° ì•¤ ë§ˆë¦¬ ì”¨(28)ì…ë‹ˆë‹¤.',
 'ì ì‹¬ì€ ì´ë ‡ê²Œ í˜¼ì ë•Œìš¸ ë•Œê°€ ë§ì•„ìš”.',
 'ê·¸ë¦¬ê³  ì˜¬ë¦¼í”½ ê¸°ë…í’ˆì¸ ë¡±íŒ¨ë”©ì´ë‚˜ ìš´ë™í™”ê°€ ì¸ê¸°ê°€ ë§ì•„ì„œ êµ¬í•˜ê¸° í˜ë“¤ë‹¤ëŠ” ì´ì•¼ê¸°ë¥¼ í–ˆëŠ”ë°, ì´ë²ˆì—ëŠ” ì´ëª¨í‹°ì½˜ì´ ì£¼ëª©ë°›ê³  ìˆìŠµë‹ˆë‹¤.'
]
```

```python
#TF-IDF 
vect = CountVectorizer()                                #CountVectorizerë¥¼ ì´ìš©í•˜ì—¬
document_term_matrix = vect.fit_transform(corpus) 
```

```python
# document_term_matrixë¥¼ dataframeìœ¼ë¡œ ë³€í™˜í•˜ê¸°!!
tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names())  
# ë³€í™˜í•œ ë‚´ìš© ì¶œë ¥í•˜ê¸°!!
tf
```

```python
D = len(tf)
df = tf.astype(bool).sum(axis=0)
idf = np.log((D+1) / (df+1)) + 1 
```

```python
df
```

```python
idf
```

```python
# TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf = tf * idf                      
tfidf = tfidf / np.linalg.norm(tfidf, axis=1, keepdims=True)
```

```python
tfidf
```

forë¬¸ì„ ì‚¬ìš©í•˜ì—¬ [n]['ko']ë¥¼ ëª¨ë‘ ë¶ˆëŸ¬ì™€ì„œ </br>

ì´ì œ CounterVectorizer() ë¥¼ ì‚¬ìš©í•´ì„œ, </br>

#### TF (Term Frequency) ë¼ëŠ” ë§ì€ ë¬¸ì„œì—ì„œ </br>
#### ì–´ë– í•œ ë‹¨ì–´(Term)ì´ ëª‡ ë²ˆì´ë‚˜, ì–¼ë§ˆë‚˜ ìì£¼ ë‚˜ì˜¤ëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì§€ìˆ˜ì…ë‹ˆë‹¤.</br>

```python
#TF-IDF 
vect = CountVectorizer()                                #CountVectorizerë¥¼ ì´ìš©í•˜ì—¬
document_term_matrix = vect.fit_transform(corpus)       # ë¬¸ì„œ-ë‹¨ì–´ í–‰ë ¬ìƒì„±!!
```

TF-IDFë¥¼ ì ìš©í–ˆìŠµë‹ˆë‹¤. </br>
ì—¬ê¸°ì„œ ìƒì„±ëœ document_term_matrixë¥¼ </br>

```python
# document_term_matrixë¥¼ dataframeìœ¼ë¡œ ë³€í™˜í•˜ê¸°!!
tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names())  
# ë³€í™˜í•œ ë‚´ìš© ì¶œë ¥í•˜ê¸°!!
tf
```

```python

D = len(tf)
df = tf.astype(bool).sum(axis=0)
idf = np.log((D+1) / (df+1)) + 1             # IDF (Inverse Document Frequency)
```

```python

# TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf = tf * idf                      
tfidf = tfidf / np.linalg.norm(tfidf, axis=1, keepdims=True)
```
